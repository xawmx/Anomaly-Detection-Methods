## 数据统计信息分析

将交易账户视为节点，账户之间的交易关系视为边。分别统计节点属性和边属性：

```python
labeled_account = list(df_label["Idx"])

# 初始化账户统计信息
account_stats = {}

# 遍历交易数据，更新统计信息
for idx, tx in df_tx.iterrows():
    sender = int(tx["From"])
    receiver = int(tx["To"])
    amount = tx["Value"]
    timestamp = int(tx["TimeStamp"])

    # 更新发送方信息
    if sender not in account_stats:
        account_stats[sender] = {"in_txs": 0, "in_amount": 0, "out_txs": 0, "out_amount": 0,\
                                 "first_active": timestamp, "last_active": timestamp}
    account_stats[sender]["out_txs"] += 1
    account_stats[sender]["out_amount"] += amount
    if timestamp < account_stats[sender]["first_active"]:
        account_stats[sender]["first_active"] = timestamp
    if timestamp > account_stats[sender]["last_active"]:
        account_stats[sender]["last_active"] = timestamp

    # 更新接收方信息
    if receiver not in account_stats:
        account_stats[receiver] = {"in_txs": 0, "in_amount": 0, "out_txs": 0, "out_amount": 0,\
                                 "first_active": timestamp, "last_active": timestamp}
    account_stats[receiver]["in_txs"] += 1
    account_stats[receiver]["in_amount"] += amount
    if timestamp < account_stats[receiver]["first_active"]:
        account_stats[receiver]["first_active"] = timestamp
    if timestamp > account_stats[receiver]["last_active"]:
        account_stats[receiver]["last_active"] = timestamp

# 计算存活周期和平均信息，打标签
for account, stats in account_stats.items():
    stats["total_txs"] = stats["in_txs"] + stats["out_txs"]
    stats["balance"] = stats["in_amount"] - stats["out_amount"]
    stats["avg_in_amount"] = stats["in_amount"] / stats["in_txs"] if stats["in_txs"] > 0 else stats["in_amount"]
    stats["avg_out_amount"] = stats["out_amount"] / stats["out_txs"] if stats["out_txs"] > 0 else stats["out_amount"]
    stats["lifespan"] = stats["last_active"] - stats["first_active"]
    stats["freq_in"] = stats["in_txs"] / (stats["lifespan"] / 86400) if stats["lifespan"] > 0 else stats["in_txs"]  # xx笔交易/天
    stats["freq_out"] = stats["out_txs"] / (stats["lifespan"] / 86400) if stats["lifespan"] > 0 else stats["out_txs"]
    stats["freq"] = stats["freq_in"] + stats["freq_out"]
    if account in labeled_account:
        stats["label"] = int(df_label[df_label["Idx"] == account]["Label"])
    else:
        stats["label"] = -1
    
    
df_account_stats = pd.DataFrame(list(account_stats.values()), index = list(account_stats.keys()))
df_account_stats = df_account_stats.drop(['first_active', 'last_active'], axis=1)
# df_account_stats.to_csv(dataset + "account_stats.csv")
df_account_stats
```

```python
# 按'From'和'To'列进行分组，并聚合Value列的和、交易数量和时间跨度
df_tx_stats = df_tx.copy()
df_tx_stats["Node_pair"] = df_tx_stats.apply(lambda row: tuple(sorted([int(row["From"]), int(row["To"])])), axis=1)
df_tx_stats = df_tx_stats.groupby("Node_pair").agg(
    Sum_amount=('Value', 'sum'),  # 将 'Value' 列求和并重命名为 'Sum_amount'
    Txs=('TimeStamp', 'count'),  # 计算 'TimeStamp' 的数量
    Time_min=('TimeStamp', 'min'),  # 计算 'TimeStamp' 的最小值
    Time_max=('TimeStamp', 'max')   # 计算 'TimeStamp' 的最大值
).rename(columns={'Value': 'Amount'}).reset_index()

df_tx_stats['TimeSpan'] = df_tx_stats['Time_max'] - df_tx_stats['Time_min']

df_tx_stats = df_tx_stats[["Node_pair", 'Sum_amount', 'Txs', 'TimeSpan']]

df_tx_stats["Avg_amount"] = df_tx_stats["Sum_amount"] / df_tx_stats["Txs"]
df_tx_stats["TimeSpan"] = df_tx_stats["TimeSpan"].replace(0, 1)  # 单位：秒

labeled_account = list(df_label["Idx"])

def construct_label(row):
    f = 0; t = 0
    if int(row["Node_pair"][0]) in labeled_account and \
    int(df_label[df_label['Idx'] == int(row["Node_pair"][0])]['Label']) == 1: # From为钓鱼账户
        f += 1
    if int(row["Node_pair"][1]) in labeled_account and \
    int(df_label[df_label['Idx'] == int(row["Node_pair"][1])]['Label']) == 1:   # To为钓鱼账户
        t += 1
    return f + t

df_tx_stats['Label'] = df_tx_stats.apply(construct_label, axis=1)
# df_tx_stats.to_csv(dataset + "tx_stats_np.csv")
df_tx_stats
```

```python
def plot_radar_chart_node(df):    
    # 获取特征列
    features = ['in_txs', 'out_txs',  'total_txs', 'in_amount', 'out_amount', 'balance', 'avg_in_amount', 'avg_out_amount', 
                'lifespan', 'freq_in', 'freq_out', 'freq']

    # 获取标签为 0 和 1 的数据
    values_0 = df.loc[0][features].values.flatten()
    values_1 = df.loc[1][features].values.flatten()

    # 计算角度
    angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()

    # 为每个特征定义刻度
    # 可以为每个轴设置不同的最大值来改变每个特征的尺度
    max_values = {
        'in_txs': 30, 'in_amount': 150, 'out_txs': 10, 'out_amount': 150, 
        'total_txs': 40, 'balance': 0.02, 'avg_in_amount': 10, 'avg_out_amount': 15, 
        'lifespan': 17000000, 'freq_in': 2, 'freq_out': 0.3, 'freq': 2.5
    }

    # 标准化处理（这里只做了简化的操作，实际可以根据需求做更复杂的变换）
    values_0 = [v / max_values[label] for v, label in zip(values_0, features)]
    values_1 = [v / max_values[label] for v, label in zip(values_1, features)]

    # 为数据添加回到起点
    values_0 += values_0[:1]
    values_1 += values_1[:1]
    angles += angles[:1]

    # 初始化雷达图
    fig, ax = plt.subplots(figsize=(4, 4), dpi=300, subplot_kw=dict(polar=True))

    # 绘制标签为 0 的数据
    ax.plot(angles, values_0, linewidth=2, linestyle='solid', label='non-phishers')
    ax.fill(angles, values_0, alpha=0.25)

    # 绘制标签为 1 的数据
    ax.plot(angles, values_1, linewidth=2, linestyle='solid', label='phishers')
    ax.fill(angles, values_1, alpha=0.25)

    # 设置雷达图标签
    ax.set_yticklabels([])  # 去除y轴标签
    ax.set_xticks(angles[:-1])  # 设置类别角度
    ax.set_xticklabels(features, fontsize=12)  # 设置类别名称

    # 自定义图表标题和图例
  #  ax.set_title('radar_chart_node_features')
    ax.legend(bbox_to_anchor=(0.8, 1), fontsize=12)

    # 保存雷达图
    plt.savefig(dataset + 'figs\\node_features_radar.png', dpi=300, bbox_inches='tight')
    plt.show()

plot_radar_chart_node(df_median[1:].copy())
```

## 图分析

```python
class tGraph(object):
    def __init__(self, df_tx, df_features, output_Gpkl=False):
        self.G = nx.MultiDiGraph() 
        self.min_time = sys.maxsize
        self.max_time = 0
        self.min_amount = sys.maxsize
        self.max_amount = 0
        
        edge_key = 0
        
        for i in df_tx.index:
            x = int(df_tx.From[i])
            y = int(df_tx.To[i])
            t = int(df_tx.TimeStamp[i])
            a = df_tx.Value[i]
            
            if not self.G.has_node(x):
                self.G.add_node(x)
                features = df_features.loc[x].values.tolist()
                self.G.nodes[x]["features"] = features[:-1]
                self.G.nodes[x]["label"] = int(features[-1])
                
            if not self.G.has_node(y):
                self.G.add_node(y)
                features = df_features.loc[y].values.tolist()
                self.G.nodes[y]["features"] = features[:-1]
                self.G.nodes[y]["label"] = int(features[-1])

            if self.G.has_edge(x,y,t):
                self.G[x][y][t]['weight'] += a                        
            else:    
                self.G.add_edge(x,y,key=t, weight=a)     

            edge_key = edge_key + 1                     
            if t < self.min_time:
                self.min_time = t
            elif t > self.max_time:
                self.max_time = t    

        self.number_of_nodes = self.G.number_of_nodes()
        self.number_of_edges = self.G.number_of_edges()
        print("Summary of graph:")
        print("Number of nodes: ", self.number_of_nodes)
        print("Number of edges: ", self.number_of_edges)
        print("Number of node features: ", len(self.G.nodes[x]["features"]))
        print("Number of edge_key: ", edge_key)
        print("Min time: ", self.min_time) 
        print("Max time: ", self.max_time) 
```

### 聚类系数

```python
# 创建一个新的有向图
simple_G = nx.DiGraph()

# 遍历 MultiDiGraph 中的节点
for node in tG.G.nodes():
    # 遍历每个节点的邻居
    for neighbor, edges in tG.G[node].items():
        # 选择权重最大的边
        max_weight = max(tG.G[node][neighbor].get('weight', 1) for _ in tG.G[node][neighbor])
        # 添加边到新的有向图
        simple_G.add_edge(node, neighbor, weight=max_weight)
    simple_G.nodes[node]["label"] = tG.G.nodes[node]["label"]

# 计算平均聚类系数
average_clustering_coefficient = nx.average_clustering(simple_G, weight='weight')
print("平均聚类系数:", average_clustering_coefficient)
```

```python
# 计算所有节点的局部聚类系数
clustering_coeffs = nx.clustering(simple_G)

nodes_n = [] # 正常节点
nodes_an = [] # 异常节点
nodes_un = [] # 未标记节点
for node in simple_G.nodes():
    if simple_G.nodes[node]['label'] == 0:
        nodes_n.append(node)
    elif simple_G.nodes[node]['label'] == 1:
        nodes_an.append(node)
    elif simple_G.nodes[node]['label'] == -1:
        nodes_un.append(node)

clustering_coeffs_n = {node: clustering_coeffs[node] for node in nodes_n}
clustering_coeffs_an = {node: clustering_coeffs[node] for node in nodes_an}
clustering_coeffs_un = {node: clustering_coeffs[node] for node in nodes_un}

# 计算异常节点聚类系数的平均值
average_clustering_coefficient_an = sum(clustering_coeffs_an.values()) / len(clustering_coeffs_an)

average_clustering_coefficient_n = sum(clustering_coeffs_n.values()) / len(clustering_coeffs_n)
average_clustering_coefficient_un = sum(clustering_coeffs_un.values()) / len(clustering_coeffs_un)

print("异常节点的平均聚类系数:", average_clustering_coefficient_an)
print("正常节点的平均聚类系数:", average_clustering_coefficient_n)
print("未标记节点的平均聚类系数:", average_clustering_coefficient_un)
```

### 特征距离

#### 全局平均特征距离（Global average distance）

衡量每对连接节点之间的平均差异

```python
global_avg_d = 0

for edge in list(tG.G.edges):
    node_from, node_to = edge[0], edge[1]
    features_node_from = tG.G.nodes[node_from]['features']
    features_node_to = tG.G.nodes[node_to]['features']
    
    global_avg_d += np.linalg.norm(np.array(features_node_from) - np.array(features_node_to))

global_avg_d = global_avg_d / tG.number_of_edges
print("Global average distance:", global_avg_d)
```

#### 局部平均特征距离（Local average distance）

衡量每个节点与其邻居之间的平均差异

```python
local_avg_d = 0

for node in list(tG.G.nodes):
    features_node = tG.G.nodes[node_from]['features']
    neighbor_d = 0
    num_neighbors = 0
    
    for neighbor in list(tG.G.successors(node)) + list(tG.G.predecessors(node)):
        features_neighbor = tG.G.nodes[neighbor]['features']
        neighbor_d += np.linalg.norm(np.array(features_node) - np.array(features_neighbor))
        num_neighbors += 1
        
    local_avg_d += (neighbor_d / num_neighbors)

local_avg_d /= tG.number_of_nodes
print("Local average distance:", local_avg_d)
```

```python
# 创建无向图对象
G = nx.Graph()

# 添加节点和边到图中，并将关系特征作为边权重
for _, row in df.iterrows():
    node1, node2 = row['Node_pair']
    features = row[['Sum_amount', 'Txs', 'TimeSpan', 'Avg_amount']].values
    G.add_edge(node1, node2, weight=features)

# 打印图的信息
print(f"Graph has {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")
```

```python
def weight_choice(normalized_probs):
    if len(normalized_probs) > 0:
        J = alias_setup(normalized_probs)[0]
        q = alias_setup(normalized_probs)[1]
        idx = alias_draw(J, q)
    return idx


def alias_setup(probs):
    K = len(probs)
    q = np.zeros(K, dtype=np.float32)
    J = np.zeros(K, dtype=np.int32)

    smaller = []
    larger = []
    for kk, prob in enumerate(probs):
        q[kk] = K * prob
        if q[kk] < 1.0:
            smaller.append(kk)
        else:
            larger.append(kk)

    while len(smaller) > 0 and len(larger) > 0:
        small = smaller.pop()
        large = larger.pop()

        J[small] = large
        q[large] = q[large] + q[small] - 1.0
        if q[large] < 1.0:
            smaller.append(large)
        else:
            larger.append(large)

    return J, q


def alias_draw(J, q):
    K = len(J)

    kk = int(np.floor(np.random.rand() * K))
    if np.random.rand() < q[kk]:
        return kk
    else:
        return J[kk]
```

```python
def get_walks(graph, num_walks, walk_length, bias=0):
    walks = []
    nodes = list(graph.nodes())
    
    # 预先计算好 alias setup 和相关信息
    alias_info = {}
    for node in nodes:
        neighbors = list(graph.neighbors(node))
        connected_features = [graph[node][neighbor]['weight'] for neighbor in neighbors]
        probs = compute_probs_with_features(connected_features, bias)
        alias_info[node] = (neighbors, probs)  # 保存每个节点的邻居和对应的采样概率

    print("Walk iteration:")
    for walk_iter in range(num_walks):
        print(f"{walk_iter + 1} / {num_walks}")
        random.shuffle(nodes)
        for node in tqdm(nodes, desc="Sampling walks", leave=False):  # 使用 leave=False 防止进度条不断刷新
            walks.append(random_walk_with_features(graph, walk_length, node, alias_info))  
            
    return walks


def random_walk_with_features(graph, walk_length, start_node, alias_info):        
    walk = [start_node]
    current_node = start_node
    
    for _ in range(walk_length):
        neighbors, probs = alias_info[current_node]  # 从预计算的信息中获取邻居和对应的概率
        
        current_node = neighbors[weight_choice(probs)]  # 根据概率进行采样
        walk.append(current_node)
        
    return walk

def compute_probs_with_features(weights, bias):
    weights = np.array(weights)

    normalized_weights = weights / weights.sum(axis=0)

    return normalized_weights[:, bias].tolist()
```

```python
num_walks = 5
walk_length = 10
feature_types = [0, 1, 2, 3]
```

```python
for bias in feature_types:
    print(f"Feature_{bias} Walking...")
    t1 = time.time()
    walks = get_walks(G, num_walks, walk_length, bias)
    t2 = time.time()
    print("Walking time:", t2-t1)

    with open(dataset + f'feature_{bias}_walks_nw{num_walks}_wl{walk_length}.pkl', 'wb') as f:
        pickle.dump(walks, f)
```

```python
def generate_embeddings_from_walks(bias, num_walks, walk_length, dimensions, window_size, workers):
    with open(dataset + f'feature_{bias}_walks_nw{num_walks}_wl{walk_length}.pkl', 'rb') as f:
        walks = pickle.load(f)
        walks = [[str(node) for node in walk] for walk in walks] #将节点编号转化为字符串

    output = dataset + f"feature_{bias}_vec_nw{num_walks}_wl{walk_length}_d{dimensions}_ws{window_size}.txt"

    print("Learn embeddings...")   
    t1 = time.time()
    word2vec_model = Word2Vec(sentences=walks, vector_size=dimensions, window=window_size, min_count=0, sg=1, hs=1, workers=workers)
    t2 = time.time()
    print("Learn embeddings time:", t2-t1) 

    vectors = {}
    for node in list(G.nodes()):
        vectors[str(node)] = word2vec_model.wv[str(node)]            

    print("Embeddings are saved in ", output)
    word2vec_model.wv.save_word2vec_format(output)
    
def load_embeddings(filename):
    fin = open(filename, 'r')
    node_num, size = [int(x) for x in fin.readline().strip().split()]
    vectors = {}
    while 1:
        l = fin.readline()
        if l == '':
            break
        vec = l.strip().split(' ')
        assert len(vec) == size + 1
        vectors[int(vec[0])] = [float(x) for x in vec[1:]]
    fin.close()
    assert len(vectors) == node_num
    return vectors
```

```python
dimensions = 16 # 嵌入维度
window_size = 4
workers =  5
```

```python
for bias in feature_types:
    generate_embeddings_from_walks(bias, num_walks, walk_length, dimensions, window_size, workers)
```

```python
def caculate_similarity(node1, node2, embeddings, method="eu"):
    # 获取节点的嵌入向量
    vec1 = embeddings.get(node1)
    vec2 = embeddings.get(node2)
    
    if method == "cos":
        return cosine_similarity([vec1], [vec2])[0][0]
    elif method == "eu":
        dist = np.linalg.norm(np.array(vec1) - np.array(vec2))
        return 1 / (1 + dist)
```

```python
for bias, col in enumerate(df.columns[1:].tolist()):
    output = dataset + f"feature_{bias}_vec_nw{num_walks}_wl{walk_length}_d{dimensions}_ws{window_size}.txt"
    embeddings = load_embeddings(output)
    
    df[col] = df['Node_pair'].apply(lambda x: caculate_similarity(x[0], x[1], embeddings, "cos"))

df.to_csv(dataset + "tx_stats.csv")
```
